```r
data <- read.csv(file="../basque_seq2seq.tsv",head=TRUE,sep="\t")
```

```r
loss<-data[,1:2]
acc<-data[,-2]
```

```r
library(ggplot2) # for plotting functions.
library(reshape2) # for melt and cast functions
```

```r
ggplot(data=loss[1:200,], aes(x=epoch*10, y=train_loss)) +
    geom_line() + 
    theme(plot.title = element_text(hjust = 0.5)) +
    labs(x="epoch",y="train_loss",title="Loss(basque)")
```

```r
error.m <- melt(acc[1:300,], id='epoch')
ggplot(data=error.m, aes(x=epoch*10, y=value, color=variable)) +
    scale_color_discrete(guide = guide_legend(title = 'Dataset')) +theme(plot.title = element_text(hjust = 0.5))+
    geom_line()+
    labs(x="epoch",y="accuracy",title="Test and Train Accuracy(basque)")
# +geom_smooth()
```
